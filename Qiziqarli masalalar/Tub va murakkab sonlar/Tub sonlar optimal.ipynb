```python
#5 Tub sonlarning optimal yechimi

n = int(input("n = "))
sanoq = 0
print(2,end=" ")
for i in range(3,n+1,2):
    k = False
    for j in range(3,int(i**(1/2))+1,2):
        sanoq += 1
        if i%j==0:
            k = True
            break
    if not k:
        print(i,end=" ")
print("Sanoq:",sanoq)
```
